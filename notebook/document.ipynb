```python
### Directory Loader
from langchain_community.document_loaders import (
    DirectoryLoader,
    PyMuPDFLoader
)

loader = DirectoryLoader(
    "../data/papers",
    glob="*.pdf",
    loader_cls=PyMuPDFLoader,
    show_progress=False
)

documents = loader.load()
###documents
print(f"Loaded {len(documents)} documents")
```
